## Ejercicio 16: DataFrame de estudiantes
**Enunciado:** Crear un DataFrame con al menos 10 estudiantes que incluya nombre, carrera, edad, promedio y ciudad. Mostrar las primeras 5 filas, la información general, el promedio general, los estudiantes con promedio mayor o igual a 80 y ordenarlos de mayor a menor promedio.

```python
# Importamos pandas para poder usar DataFrames
import pandas as pd

# Creamos los datos de los estudiantes en un diccionario
datos = {
    "nombre":   ["Ana López", "Carlos Pérez", "María García", "José Martínez", "Laura Díaz",
                 "Pedro Ramírez", "Sofía Torres", "Luis Hernández", "Valeria Cruz", "Diego Flores"],
    "carrera":  ["Sistemas", "Contabilidad", "Derecho", "Medicina", "Sistemas",
                 "Derecho", "Medicina", "Contabilidad", "Sistemas", "Medicina"],
    "edad":     [20, 22, 21, 23, 20, 24, 21, 22, 20, 23],
    "promedio": [85, 72, 90, 65, 88, 78, 95, 70, 83, 60],
    "ciudad":   ["Tegucigalpa", "San Pedro Sula", "Tegucigalpa", "La Ceiba", "Comayagua",
                 "San Pedro Sula", "Tegucigalpa", "La Ceiba", "Comayagua", "Tegucigalpa"]
}

# Convertimos el diccionario en un DataFrame
df = pd.DataFrame(datos)

# Mostramos las primeras 5 filas
print("Primeras 5 filas:")
print(df.head())

# Mostramos la información general del DataFrame
print("\nInformación general:")
print(df.info())

# Calculamos el promedio general de todos los estudiantes
promedio_general = df["promedio"].mean()
print("\nPromedio general de calificaciones:", promedio_general)

# Mostramos solo los estudiantes con promedio mayor o igual a 80
print("\nEstudiantes con promedio mayor o igual a 80:")
print(df[df["promedio"] >= 80])

# Ordenamos de mayor a menor promedio
print("\nEstudiantes ordenados por promedio (mayor a menor):")
print(df.sort_values("promedio", ascending=False))
```

**Explicación de métodos pandas:**
- `pd.DataFrame()` convierte un diccionario en una tabla organizada en filas y columnas.
- `head()` muestra solo las primeras 5 filas, para revisar rápido cómo quedaron los datos.
- `info()` muestra un resumen del DataFrame: cuántas filas hay, qué tipo de dato tiene cada columna y si hay valores vacíos.
- `mean()` calcula el promedio de todos los valores de una columna.
- `df[df["promedio"] >= 80]` filtra la tabla y muestra solo las filas que cumplen esa condición.
- `sort_values()` ordena la tabla según una columna. Con `ascending=False` ordena de mayor a menor.

## Ejercicio 17: Análisis de ventas con pandas 
**Enunciado:**
- Cree un DataFrame de ventas con al menos 15 registros. 
- Incluya fecha, vendedor, producto, categoría, cantidad y precio unitario. 
- Calcule una nueva columna TotalVenta. 
- Calcule el total vendido por vendedor. 
- Identifique el producto más vendido. 
- Identifique la categoría con mayor ingreso. 
- Ordene las ventas de mayor a menor.

**Explicación de la lógica:**
- pd.DataFrame(datos)  crea una tabla de datos usando pandas.
- ventas["TotalVenta"] crea una nueva columna multiplicando cantidad por precio unitario.
- groupby() agrupa datos por una columna.
- sum() suma valores numéricos.
- idxmax() devuelve el elemento con el valor máximo.
- sort_values() ordena los datos.
ascending=False ordena de mayor a menor.

```python
import pandas as pd

# Datos de ventas
datos = {
    "Fecha": [
        "2025-05-01","2025-05-02","2025-05-03","2025-05-04","2025-05-05",
        "2025-05-06","2025-05-07","2025-05-08","2025-05-09","2025-05-10",
        "2025-05-11","2025-05-12","2025-05-13","2025-05-14","2025-05-15"
    ],

    "Vendedor": [
        "Ana","Luis","Pedro","Ana","Luis",
        "Pedro","Ana","Luis","Pedro","Ana",
        "Luis","Pedro","Ana","Luis","Pedro"
    ],

    "Producto": [
        "Laptop","Mouse","Teclado","Monitor","Laptop",
        "Mouse","Teclado","Laptop","Monitor","Mouse",
        "Laptop","Teclado","Monitor","Mouse","Laptop"
    ],

    "Categoria": [
        "Tecnologia","Accesorios","Accesorios","Tecnologia","Tecnologia",
        "Accesorios","Accesorios","Tecnologia","Tecnologia","Accesorios",
        "Tecnologia","Accesorios","Tecnologia","Accesorios","Tecnologia"
    ],

    "Cantidad": [
        2,5,3,1,2,
        4,2,1,2,3,
        2,4,1,6,3
    ],

    "PrecioUnitario": [
        800,20,35,250,800,
        20,35,800,250,20,
        800,35,250,20,800
    ]
}

# Crear DataFrame
ventas = pd.DataFrame(datos)

# Nueva columna TotalVenta
ventas["TotalVenta"] = ventas["Cantidad"] * ventas["PrecioUnitario"]

print("DataFrame completo")
print(ventas)

# Total vendido por vendedor
total_vendedor = ventas.groupby("Vendedor")["TotalVenta"].sum()

print("\nTotal vendido por vendedor")
print(total_vendedor)

# Producto más vendido
producto_mas_vendido = ventas.groupby("Producto")["Cantidad"].sum()

print("\nProducto más vendido")
print(producto_mas_vendido.idxmax())

# Categoría con mayor ingreso
categoria_ingreso = ventas.groupby("Categoria")["TotalVenta"].sum()

print("\nCategoría con mayor ingreso")
print(categoria_ingreso.idxmax())

# Ordenar ventas
ventas_ordenadas = ventas.sort_values(
    by="TotalVenta",
    ascending=False
)

print("\nVentas ordenadas")
print(ventas_ordenadas)
```

## Ejercicio 18: Limpieza de datos con pandas
Enunciado: Crear un DataFrame con datos incompletos de clientes. Identificar valores nulos, contar nulos por columna, reemplazar ciudades vacías y eliminar registros sin correo.

Lógica: Se usan listas y ciclos para revisar datos vacíos. Luego se reemplazan ciudades vacías y se eliminan clientes que no tienen correo.

```python
clientes = [
    ["Ana", 22, "ana@gmail.com", "Tegucigalpa", "9999-1111"],
    ["Carlos", 35, "", "", "8888-2222"],
    ["María", "", "maria@gmail.com", "San Pedro Sula", ""],
    ["José", 28, "jose@gmail.com", "", "7777-3333"],
    ["Lucía", 40, "", "La Ceiba", "6666-4444"]
]

print("DATOS ORIGINALES")
for cliente in clientes:
    print(cliente)

# Contar valores vacíos
nulos = 0

for cliente in clientes:
    for dato in cliente:
        if dato == "":
            nulos += 1

print("\nCantidad de valores vacíos:", nulos)

# Reemplazar ciudades vacías
for cliente in clientes:
    if cliente[3] == "":
        cliente[3] = "No especificada"

# Eliminar clientes sin correo
clientes_limpios = []

for cliente in clientes:
    if cliente[2] != "":
        clientes_limpios.append(cliente)

print("\nDATOS LIMPIOS")

for cliente in clientes_limpios:
    print(cliente)
```

# Ejercicio 19: Análisis de préstamos

**Enunciado:**  
Crear un DataFrame con solicitudes de préstamo que incluya cliente, monto solicitado, plazo, ingreso mensual y estado de solicitud. Calcular el monto promedio solicitado, contar solicitudes por estado, mostrar únicamente préstamos aprobados, clientes con monto mayor a 50,000 y agrupar por estado calculando el promedio del monto solicitado.

```python
# Importamos pandas
import pandas as pd

# Datos de los prestamos
datos = {
    "cliente": ["Ana Lopez", "Carlos Perez", "Maria Gomez", "Jose Ruiz",
                "Lucia Torres", "Pedro Martinez", "Sofia Castro", "Luis Herrera"],

    "monto_solicitado": [45000, 60000, 30000, 80000,
                         52000, 47000, 90000, 25000],

    "plazo": [12, 24, 18, 36, 24, 12, 48, 6],

    "ingreso_mensual": [15000, 25000, 12000, 40000,
                        22000, 18000, 50000, 10000],

    "estado": ["Aprobado", "Pendiente", "Rechazado", "Aprobado",
               "Aprobado", "Pendiente", "Aprobado", "Rechazado"]
} 

# Crear el DataFrame
df = pd.DataFrame(datos)

# Mostrar las primeras filas
print("Primeras 5 filas:")
print(df.head())

# Informacion general
print("\nInformacion general:")
print(df.info())

# Promedio del monto solicitado
promedio = df["monto_solicitado"].mean()

print("\nPromedio del monto solicitado:", promedio)

# Cantidad de solicitudes por estado
print("\nSolicitudes por estado:")
print(df["estado"].value_counts())

# Solo prestamos aprobados
print("\nPrestamos aprobados:")
print(df[df["estado"] == "Aprobado"])

# Clientes con monto mayor a 50000
print("\nClientes con monto mayor a 50000:")
print(df[df["monto_solicitado"] > 50000])

# Agrupar por estado y sacar promedio
print("\nPromedio por estado:")
print(df.groupby("estado")["monto_solicitado"].mean())
```

# Explicación de métodos pandas:

- `pd.DataFrame()` convierte el diccionario en una tabla organizada.
- `head()` muestra las primeras 5 filas.
- `info()` muestra información general del DataFrame.
- `mean()` calcula el promedio del monto solicitado.
- `value_counts()` cuenta solicitudes por estado.
- `groupby()` agrupa los datos por estado.

# Ejercicio 20: Mini análisis de datos tipo empresa

**Enunciado:**  
Crear un DataFrame con información de empleados que incluya nombre, departamento, cargo, salario, años de experiencia y ciudad. Mostrar los primeros registros, estadísticas generales de salarios, calcular el salario promedio por departamento, encontrar el empleado con salario más alto, mostrar empleados con más de 5 años de experiencia, ordenar empleados por salario y crear una columna llamada NivelExperiencia con Junior, Intermedio o Senior.

```python
# ==========================================
# Ejercicio 20: Mini análisis de datos tipo empresa
# ==========================================

# importar pandas
import pandas as pd

# crear los datos de empleados
datos = {
    "Nombre": ["Juan", "Maria", "Carlos", "Ana", "Luis", "Sofia"],
    "Departamento": ["TI", "Ventas", "Recursos Humanos", "TI", "Ventas", "Finanzas"],
    "Cargo": ["Programador", "Vendedor", "Analista", "Administrador", "Supervisor", "Contadora"],
    "Salario": [25000, 18000, 22000, 30000, 27000, 24000],
    "Experiencia": [2, 6, 4, 8, 10, 5],
    "Ciudad": ["Tegucigalpa", "San Pedro Sula", "La Ceiba", "Tegucigalpa", "Choloma", "Comayagua"]
}

# convertir a DataFrame
df = pd.DataFrame(datos)

# mostrar los primeros registros
print("Primeros registros")
print(df.head())

# estadisticas de salarios
print("\nEstadisticas de salarios")
print(df["Salario"].describe())

# promedio de salario por departamento
print("\nPromedio de salario por departamento")
print(df.groupby("Departamento")["Salario"].mean())

# empleado con salario mas alto
print("\nEmpleado con salario mas alto")
print(df[df["Salario"] == df["Salario"].max()])

# empleados con mas de 5 años de experiencia
print("\nEmpleados con mas de 5 años de experiencia")
print(df[df["Experiencia"] > 5])

# ordenar empleados por salario
print("\nEmpleados ordenados por salario")
print(df.sort_values(by="Salario", ascending=False))

# crear columna nivel de experiencia
niveles = []

for años in df["Experiencia"]:
    
    if años <= 3:
        niveles.append("Junior")
        
    elif años <= 7:
        niveles.append("Intermedio")
        
    else:
        niveles.append("Senior")

df["NivelExperiencia"] = niveles

# mostrar dataframe final
print("\nDataFrame final")
print(df)
```
